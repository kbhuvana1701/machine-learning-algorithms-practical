```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv(r"C:\Users\nirma\OneDrive\Documents\MELBOURNE_HOUSE_PRICES_LESS.csv (2)\MELBOURNE_HOUSE_PRICES_LESS.csv")
dataset.head()
```

```python
dataset.nunique()
```

```python
# let's use limited columns which makes more sense for serving ourpurpose
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG','Regionname', 'Propertycount',
'Distance', 'CouncilArea','Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
#Checking for Nan values
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance',]
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset['Price'] =dataset['Price'].fillna(dataset.Price.mean())
```

```python
dataset.dropna(inplace=True)
dataset.shape
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y,
test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
ridge_reg.score(train_X, train_y)
```
