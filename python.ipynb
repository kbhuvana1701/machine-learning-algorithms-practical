```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
print("\nUsing s1 =", s1, "and s2 =", s2)
```

```python
# 7. Union - all elements from both sets
print("Union s1 | s2:", s1 | s2)
```

```python
# 8. Intersection - common elements
print("Intersection s1 & s2:", s1 & s2)
```

```python
# 9. Difference - elements in s1 but not in s2
print("Difference s1 - s2:", s1 - s2)
```

```python
# 10. Symmetric difference - elements in either set but not both
print("Symmetric difference s1 ^ s2:", s1 ^ s2) 
```

```python
# 11. Check membership
print("3 in s1:", 3 in s1) 
```

```python
# 12. Iterate set
print("Iterating s1:")
for i in s1:
    print(i)
```

```python
# 1. Create dictionary
d = {'a': 1, 'b': 2}
print("Initial dict:", d)
```

```python
# 2. Access value
print("d['a'] =", d['a'])
```

```python
# 3. Add key-value pair
d['c'] = 3
print("After d['c'] = 3:", d)
```

```python
# 4. Update value
d['a'] = 10
print("After d['a'] = 10:", d)
```

```python
# 5. Remove key - returns the value
removed = d.pop('b')
print("d.pop('b') returned:", removed) 
```

```python
# 6. Get keys
print("d.keys() =", d.keys()) 
```

```python
# 7. Get values
print("d.values() =", d.values()) 
```

```python
# 8. Iterate keys
print("Iterating keys:")
for k in d:
    print(k)
```

```python
# 9. Iterate items
print("Iterating items:")
for k, v in d.items():
    print(k, v)
```

```python
# 10. Check membership - checks keys only
print("'a' in d =", 'a' in d)
```

```python
# 11. Length
print(len(d)) 
```

```python
# 12. Clear dictionary
s=d.clear()
print(s)
```
